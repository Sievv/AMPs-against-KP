```python
!pip install -U imbalanced-learn
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.metrics import (
    confusion_matrix,
    matthews_corrcoef,
    roc_auc_score,
    average_precision_score,
    precision_recall_curve,
    roc_curve,
    recall_score
)
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier

# Load the resampled data (already done after applying SMOTE)
X_resampled = pd.read_csv('/content/drive/MyDrive/KP/KP smote/X_resampled.csv')  # Replace with the actual path for X_resampled
y_resampled = pd.read_csv('/content/drive/MyDrive/KP/KP smote/y_resampled.csv')  # Replace with the actual path for y_resampled

# Display the first few rows to confirm
print(X_resampled.head())
print(y_resampled.head())

# Define the models you want to evaluate
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(eval_metric='logloss', random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(n_estimators=100, random_state=42),
    'MLP': MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42),
    'SVM': SVC(probability=True, random_state=42),  # Adding SVM
}

# Prepare lists to store results
f1_scores = []
roc_aucs = []
pr_aucs = []
sensitivities = []
specificities = []
mcc_scores = []

# Calculate the number of rows needed for the subplots
n_rows = (len(models) + 1) // 2  # Integer division rounds down, add 1 before division to round up if needed
n_cols = 2

# Plot the confusion matrices in a publication style
plt.figure(figsize=(12, n_rows * 5))  # Adjust figure height based on number of rows

# Train, evaluate and collect metrics for each model
for i, (name, model) in enumerate(models.items(), 1):
    print(f"Evaluating {name}...")

    # Cross-validation: F1 score
    f1_cv_scores = cross_val_score(model, X_resampled, y_resampled, cv=5, scoring='f1')
    mean_f1 = np.mean(f1_cv_scores)
    std_f1 = np.std(f1_cv_scores)

    # Train the model on the entire resampled training set for later evaluation
    model.fit(X_resampled, y_resampled)

    # ROC AUC Score
    y_prob = model.predict_proba(X_resampled)[:, 1]  # Probability for positive class
    roc_auc = roc_auc_score(y_resampled, y_prob)

    # PR AUC Score
    pr_auc = average_precision_score(y_resampled, y_prob)

    # Store results for comparison
    f1_scores.append((name, mean_f1, std_f1))
    roc_aucs.append((name, roc_auc))
    pr_aucs.append((name, pr_auc))

    # Confusion Matrix
    y_pred = model.predict(X_resampled)
    cm = confusion_matrix(y_resampled, y_pred)
    tn, fp, fn, tp = cm.ravel()

    # Calculate Sensitivity, Specificity, and MCC
    sensitivity = recall_score(y_resampled, y_pred)  # True Positive Rate
    specificity = tn / (tn + fp)  # True Negative Rate
    mcc = matthews_corrcoef(y_resampled, y_pred)

    # Store additional metrics
    sensitivities.append((name, sensitivity))
    specificities.append((name, specificity))
    mcc_scores.append((name, mcc))

    # Display metrics
    print(f"{name} - Mean F1 Score: {mean_f1:.4f} (+/- {std_f1:.4f})")
    print(f"{name} - ROC AUC: {roc_auc:.4f}")
    print(f"{name} - PR AUC: {pr_auc:.4f}")
    print(f"{name} - Sensitivity: {sensitivity:.4f}")
    print(f"{name} - Specificity: {specificity:.4f}")
    print(f"{name} - MCC: {mcc:.4f}")

    # Plot confusion matrix
    plt.subplot(n_rows, n_cols, i)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Pred: Inactive', 'Pred: Active'],
                yticklabels=['True: Inactive', 'True: Active'])
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel('Predicted')
    plt.ylabel('True')

plt.tight_layout()
plt.show()

# Summary of model comparison
print("\nModel Comparison Summary:")

# F1 Scores
print("\nF1 Scores:")
for model_name, f1, std in f1_scores:
    print(f"{model_name}: {f1:.4f} (+/- {std:.4f})")

# ROC AUC Scores
print("\nROC AUC Scores:")
for model_name, auc in roc_aucs:
    print(f"{model_name}: {auc:.4f}")

# PR AUC Scores
print("\nPR AUC Scores:")
for model_name, auc in pr_aucs:
    print(f"{model_name}: {auc:.4f}")

# Sensitivity
print("\nSensitivity (Recall) Scores:")
for model_name, sensitivity in sensitivities:
    print(f"{model_name}: {sensitivity:.4f}")

# Specificity
print("\nSpecificity Scores:")
for model_name, specificity in specificities:
    print(f"{model_name}: {specificity:.4f}")

# MCC Scores
print("\nMCC Scores:")
for model_name, mcc in mcc_scores:
    print(f"{model_name}: {mcc:.4f}")

# Optional: If you want to visualize PR curves for all models
plt.figure(figsize=(12, 5))
for i, (name, model) in enumerate(models.items(), 1):
    # Calculate Precision-Recall curve
    precision, recall, _ = precision_recall_curve(y_resampled, model.predict_proba(X_resampled)[:, 1])
    plt.plot(recall, precision, label=f'{name} (PR AUC = {pr_aucs[i-1][1]:.4f})')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve Comparison')
plt.legend(loc='lower left')
plt.show()
```

```python
from lime.lime_tabular import LimeTabularExplainer
import joblib
import pandas as pd
import matplotlib.pyplot as plt

# Load training data (unscaled original data)
X_train = pd.read_csv('/content/drive/MyDrive/KP/KP smote/X_resampled.csv')
feature_names = X_train.columns.tolist()

# Load test data
X_test = pd.read_csv('/content/drive/MyDrive/KP/KP smote/X_test.csv')
y_test = pd.read_csv('/content/drive/MyDrive/KP/KP smote/y_test.csv').squeeze()

# Drop 'Sequence' column if present
if 'Sequence' in X_test.columns:
    X_test_numeric = X_test.drop(columns=['Sequence'])
else:
    X_test_numeric = X_test.copy()

# Choose an active instance (label = 1) from the test set
active_instances = X_test_numeric[y_test == 1]
if active_instances.empty:
    # fallback to first instance if no active found
    instance_to_explain = X_test_numeric.iloc[0].values
else:
    instance_to_explain = active_instances.iloc[0].values

# Initialize LIME explainer
explainer = LimeTabularExplainer(
    training_data=X_train.values,
    feature_names=feature_names,
    class_names=['Inactive', 'Active'],
    mode='classification'
)

# Define top 3 models and their paths
models = {
    'Random Forest': '/content/drive/MyDrive/KP/KP smote/Random Forest_model.pkl',
    'Gradient Boosting': '/content/drive/MyDrive/KP/KP smote/Gradient Boosting_model.pkl',
    'AdaBoost': '/content/drive/MyDrive/KP/KP smote/AdaBoost_model.pkl'
}

# Explain and plot feature importance for each model
for name, path in models.items():
    print(f"\nExplaining with {name} model:")
    model = joblib.load(path)

    explanation = explainer.explain_instance(
        instance_to_explain,
        model.predict_proba,
        num_features=10
    )

    print(f"Predicted probabilities ({name}): {model.predict_proba(instance_to_explain.reshape(1, -1))[0]}")

    explanation.as_pyplot_figure()
    plt.title(f"LIME Explanation for {name}")
    plt.tight_layout()
    plt.show()
```
